```python
root_path = 'miniconda3'
import sys
sys.path.append(f'{root_path}envs/mp/lib/python3.12.4/site-packages/')

import meep as mp
import numpy as np
import matplotlib.pyplot as plt
from mpi4py.MPI import COMM_WORLD
from IPython.display import clear_output, Math, HTML
```

```python
WvlenArray = [3,2.9,2.8,2.7]

# for 80um waveguide --------------------------------------
SeriesSim = [473.4,405,389.1,368.9]  # sum = 1636.4
ParallelSim = [1365]

# for 40um waveguide --------------------------------------
SeriesSim3 = [236.4,200.2,196.2,200.2]  # sum = 833
ParallelSim3 = [605.75]

# for 20um waveguide --------------------------------------
SeriesSim2 = [104.1,108.1,132.2,132.1]  # sum = 476.5
ParallelSim2 = [344.9]


print(np.sum(SeriesSim3))
```

```python
plt.figure(figsize=(8,6))
plt.plot([20,40,80],[1636.4,833,476.5],'-o',label='Series Simulation')
plt.plot([20,40,80],[1365,605.75,344.9],'-o',label='Parallel Simulation')
plt.ylabel('Simulation time (seconds)',fontsize=12)
plt.xlabel('Waveguide length (um)',fontsize=12)
plt.title('Simulation time x Waveguide length',fontsize=12)
plt.legend(fontsize=12)
plt.grid()
```

```python
plt.figure(figsize=(8,6))
plt.plot([20,40,80],[1636.4/1365,833/605.75,476.5/344.9],'-o',label='Ratio (Series/Parallel)')
plt.ylabel('Ratio (Series/Parallel) in simulation time',fontsize=12)
plt.xlabel('Waveguide length (um)',fontsize=12)
plt.title('Ratio between Series and Parallel Simulation times',fontsize=12)
plt.legend(fontsize=12)
plt.grid()
```

```python
plt.figure(figsize=(8,6))
plt.plot(WvlenArray,SeriesSim,'-o',label='80um waveguide')
plt.plot(WvlenArray,SeriesSim3,'-o',label='40um waveguide')
plt.plot(WvlenArray,SeriesSim2,'-o',label='20um waveguide')
plt.ylabel('Simulation time (seconds)',fontsize=12)
plt.xlabel('Wavelength (um)',fontsize=12)
plt.title('Simulation time x Wavelength (in Series simulation approach)',fontsize=14)
plt.legend(fontsize=12)
plt.grid()
```
